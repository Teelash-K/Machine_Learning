```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score as r2
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
from sklearn.preprocessing import MinMaxScaler
normalizer = MinMaxScaler()
from sklearn.preprocessing import StandardScaler
normalizer = StandardScaler()
```

```python
data = pd.read_csv('dataset\Loan_Data.csv')
data
```

```python
loan = data.copy()
```

```python
loan.isnull().sum()
```

```python
loan.drop(['Loan_ID'], axis = 1, inplace = True )
```

```python
loan.Dependents.sample().values
```

```python
# Number of Dependents is seen as categorical
# loan.Dependents.sample(200).values
loan['Dependents'] = loan['Dependents'].str.replace('+','')
loan['Dependents'] = loan['Dependents'].astype(float)
loan['Dependents'].dtypes
```

```python
# Check if the changes has been made to the 'Number of Dependent Variable'
cat = data.select_dtypes(include = ['category', 'object'])
num = data.select_dtypes(include = 'number')
display(cat.head())
display(num.head())
```

```python
# ---------- Clean The Dataset -----------
for i in loan.columns:
    if loan[i].dtypes == 'O':
        loan[i].fillna(loan[i].mode()[0], inplace = True)
    elif loan[i].dtypes == 'float':
        loan[i].fillna(loan[i].median(), inplace = True)
    else:
        loan[i].fillna(loan[i].median(), inplace = True)

loan.isnull().sum().sort_values(ascending = False).head(2)
```
